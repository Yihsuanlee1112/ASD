```python
import numpy as np
from numpy import linalg as LA
import os
import pandas as pd
```

```python
def calFeatures(data, interferenceType, remove=False):
    
    if interferenceType == "Inf":
        data = data[data["GlobalData._stage"] != "None"]
    elif interferenceType == "NoInf":
        data = data[data["GlobalData._stage"] == "None"]
    elif interferenceType == "All":
        pass
    
    if remove:
        data[data == 0] = np.nan
        data[data == 1] = np.nan
    
    result = pd.DataFrame()

    column_name = ['Asymmetry', 'AttentionData', 'LeftActivityData', 'Relaxation', 'RightActivityData']
    featureList = []    
    for col in column_name:
               
        featureList.append(np.nanmean(data[col]))
        featureList.append(np.nanstd(data[col]))
        
        dfFea = pd.DataFrame([featureList], columns = [f"{interferenceType}_{col}_mean", f"{interferenceType}_{col}_std"])
        result = pd.concat([result, dfFea], axis = 1)
        featureList = []
    
    
    return result
```

```python
def calMindFeatures(data):
    
    stageTypes = ["Inf","NoInf","All"]
    
    result = pd.DataFrame()
    
    for stageType in stageTypes:    
        #dfFeatures = calFeatures(data, stageType, True)
        dfFeatures = calFeatures(data, stageType)
        
        result = pd.concat([result, dfFeatures], axis = 1)
        
    return result
```

```python
def readCSV(path):
    try:
        df = pd.read_csv(path)
    except:
        print("[File not file] "+path)
        df = pd.DataFrame()
    return df
```

```python
rawDataPath = "C:\\Users\\wmlab\\Desktop\\萌芽計畫\\Model2 Data\\Rebuild\\children1"
featureType = ["EyeFocusData"]
testType = ["c","a","s"]

result = pd.DataFrame()
user = []

result = pd.DataFrame()

for folder in os.listdir(rawDataPath): # read folders    
    if folder[-1] == testType[1]:
        print(folder)
        try:
            data = pd.read_csv(os.path.join(rawDataPath,folder,"Mindfrog.EegDeviceData.EEGMindIndexesData.csv"))
            
            MindFeatures = calMindFeatures(data)
            result = pd.concat([result, MindFeatures], axis = 0)
            
            user.append(folder.split("-")[0])
        except:
            print(folder.split("-")[0] +" not found")
```
